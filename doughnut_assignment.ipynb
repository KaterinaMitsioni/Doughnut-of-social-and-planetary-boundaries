### Ecological Overshoot

Παρατηρούμε στην σελίδα 8 του άρθρου ότι για κάθε dimension kai indicator υπάρχει ένα κατώτατο όριο το οποίο βρίσκεται στο αρχείο που διαβάζεται παρακάτω μαζί με το αντίστοιχο ratio του. Στην περίπτωση των eco dimensions η διαφορά ratio - 1 δείχνει το overshoot, δηλαδή πόσο παραπάνω ξεφεύγουμε 
από τον εκάστοτε δείκτη. Επιπλέον, σύμφωνα με το άρθρο κρατάμε μόνο το έτος 2017.

```python
boundaries_df = pd.read_csv('https://raw.githubusercontent.com/a-fanning/doughnut-v3/main/Analysis-Final/myData/5_20250108_doughnutDat-boundariesRatios.csv')
```

Αφού θέλουμε trellis plot, συγκεντρώνουμε τους απαιτούμενους δείκτες σε λίστα και δημιουργούμε sublots με καθέναν από αυτους.
Για καλύτερο αισθητικό αποτέλεσμα επεξεργαζόμαστε το κάθε διάγραμμα ανάλογα με την θέση του (απόκρυψη τιμών αξόνβν κτλ)

```python
import pandas as pd
import matplotlib.pyplot as plt

eco_indicators = [
    ("blueH2O_footprint", "Blue Water footprint"),
    ("co2_footprint", "Carbon footprint"),
    ("hanpp_footprint", "HANPP footprint"),
    ("N_footprint", "Nitrogen footprint"),
    ("P_footprint", "Phosphorus footprint"),
    ("species_footprint", "Species lost footprint")
]

groups = ['Bottom-40', 'Middle-40', 'Top-20']
colors = ['#ffd300', '#cdc3ea', '#1c649c']

fig, axes = plt.subplots(2, 3, figsize=(9, 8))
axes = axes.flatten()

for i, (indicator, title) in enumerate(eco_indicators):
    ax = axes[i]

    if indicator in ['blueH2O_footprint', 'species_footprint', 'hanpp_footprint', 'co2_footprint']:
        df = df_biodiversity
    else:
        df = df_nutrientPollution

    filtered = boundaries_df[
        (boundaries_df['indicator'] == indicator) &
        (boundaries_df['group'].isin(groups)) &
        (boundaries_df['date'] == 2017)
    ]

    group_ratios = filtered.groupby('group')['ratio'].mean().reindex(groups, fill_value=0)
    group_overshoot_pct = ((group_ratios - 1).clip(lower=0)) * 100

    bars = ax.bar(groups, group_overshoot_pct, color=colors, width=0.9)

    ax.set_title(title, fontsize=12)
    if i in [0, 3]: 
        ax.set_ylabel('Ecological overshoot', fontweight='bold')

    ax.set_ylim(0, 1000)
    ax.set_yticks([0, 200, 400, 600, 800, 1000])
    ax.set_yticklabels(['0%', '200%', '400%', '600%', '800%', '1000%'])
    ax.set_xticks([])

    if i % 3 != 0:
        ax.set_yticks([])

for j in range(3):
    ax_top = axes[j]
    secax = ax_top.secondary_xaxis('top')
    secax.set_xlim(ax_top.get_xlim())
    secax.set_xticks(range(len(groups)))
    secax.set_xticklabels(groups, rotation=40, ha='left', fontsize=9)
    secax.tick_params(axis='x', pad=5)
    secax.spines['top'].set_color('0.7')
    secax.spines['top'].set_linewidth(0.5)
#axnoi axones
for ax in axes:
    for side in ['left', 'right', 'top', 'bottom']:
        ax.spines[side].set_color('0.7')     # ανοιχτό γκρι
        ax.spines[side].set_linewidth(0.5)   # λεπτό
#aristeroi axones y
strong_idxs = [0, 3]
for idx in strong_idxs:
    ax = axes[idx]
    ax.spines['left'].set_color('0.6') 
    ax.spines['left'].set_linewidth(1.5) 
plt.tight_layout()
plt.show()
```

(σελ. 8).
Μια διαφορά είναι ότι οι οικολογικοί δείκτες εκφράζουν πόσο υπολειπόμαστε απο το όριο άρα το shortfall το εκφράζει η διαφορά 1 - ratio άρα θα χρησιμοποιήσουμε

Oρίζουμε την χρωματική παλέτα, κάνουμε το mapping που θα εμφανίζεται στο hover και υπολογίζουμε τις ακτίνες και τα μήκη των ¨φετών¨ του ντόνατ. Επίσης λαμβάνουμε υπόψιν το overshoot που θα δείξει ποιοι δείκτες αυξήθηκαν δραματικά και φαίνονται με σκουρότερο χρώμα.

```python
# ECO
r_in_eco = 2.1
r_max_eco = 3.4
r_max_extra = 4.6
threshold = 1.0

v_norm = (v_raw - v_raw.min()) / (v_raw.max() - v_raw.min() + 1e-9)

base_width = 360 / n_eco
k = 0.5
widths = base_width * (1 + k * (v_norm - v_norm.mean()))
widths = widths * 360 / widths.sum()

offset_eco = 90 
theta_start = offset_eco - np.concatenate(([0], np.cumsum(widths[:-1])))
theta_end   = offset_eco - np.cumsum(widths)

eco_colors = [
    "#4a202a",  # πιο σκούρο (deep wine)
    "#713040",
    "#8f386d",
    "#9b314c",
    "#a4407c",
    "#ea87a0",
    "#f4a0b7"   # πιο ανοιχτό
]
n_colors = len(eco_colors)

def value_to_color(vn, code=None):
    idx = int((1 - vn) * (n_colors - 1) + 1e-9)
    idx = max(0, min(n_colors - 1, idx))

    if code == "CP1": #xeirokinita gia pio entoni diafora
        idx = 1

    return eco_colors[idx]


soc_to_color = value_to_color

eco_titles = {
    "CC1": "CO₂ concentration",
    "CC2": "Radiative forcing",
    "OA1": "Ocean acidification",
    "CP1": "Hazardous chemicals production",
    "NP1": "Phosphorus pollution",
    "NP2": "Nitrogen pollution",
    "AP1": "Air pollution",
    "FD1": "Blue water flows",
    "FD2": "Soil moisture",
    "LC1": "Forest area",
    "BB1": "Species extinctions",
    "BB2": "Human appropriation of NPP",
    "OD1": "Ozone depletion"
}

social_titles_by_indicator = {
    "undernourishment": "Undernourishment",
    "foodInsecurity": "Food insecurity",
    "under5death": "Child mortality",
    "UHCindex": "Lack of health services",
    "adultLiteracy": "Illiteracy",
    "secondarySchool": "Incomplete secondary school",
    "societalPoverty": "Societal poverty",
    "youthNEET": "Youth unemployment",
    "drinkingH2O": "Unsafe drinking water",
    "sanitation": "Unsafe sanitation",
    "energyAccess": "Lack of electricity",
    "energyIndoor": "Lack of clean fuels indoors",
    "publicTrans": "Lack of public transport",
    "internet": "Lack of internet",
    "urbanSlums": "Inadequate housing",
    "genderGapIndex": "Gender inequality",
    "socialSupport": "Lack of social support",
    "palma": "Income inequality",
    "govRegimes": "Autocratic regimes",
    "controlCorruption": "Perceptions of corruption",
    "homicideOver5": "Homicides"
}

# gia eco overshoot
v_base = np.minimum(v_raw, threshold)
v_extra = np.clip(v_raw - threshold, 0, None)

v_base_norm = v_base / (threshold + 1e-9)
v_base_sqrt = np.sqrt(v_base_norm)
alpha = 0.6
v_base_scaled = alpha * v_base_sqrt + (1 - alpha) * 1.0
v_base_scaled = (v_base_scaled - v_base_scaled.min()) / (v_base_scaled.max() - v_base_scaled.min() + 1e-9)

r_in_arr_eco = np.full_like(v_base_scaled, r_in_eco)
r_out_base = np.sqrt(r_in_eco**2 + v_base_scaled * (r_max_eco**2 - r_in_eco**2))

v_extra_norm = np.zeros_like(v_extra)
mask_extra   = v_extra > 0
if mask_extra.any():
    v_extra_norm[mask_extra] = v_extra[mask_extra] / (v_extra[mask_extra].max() + 1e-9)
r_out_extra = r_out_base + v_extra_norm * (r_max_extra - r_max_eco)

# SOCIAL
r_out_social_ring = 2.4
r_in_social_ring  = 1.6

v_soc = soc_valid["v"].values.clip(min=0)
has_100 = np.isclose(v_soc, 100).any()
if has_100:
    v_soc_norm = (v_soc - v_soc.min()) / (v_soc.max() - v_soc.min() + 1e-9)
else:
    v_soc_norm = np.clip(v_soc / 100.0, 0, 1)

r_outer_soc = r_in_social_ring
r_min_soc   = 0.1

r_span = r_outer_soc - r_min_soc
r_in_arr_soc = r_outer_soc - v_soc_norm * r_span * 0.9
r_in_arr_soc = np.clip(r_in_arr_soc, r_min_soc, r_outer_soc - 0.05)
r_out_arr_soc = np.full_like(r_in_arr_soc, r_outer_soc)

offset_soc = 90
angles_soc = offset_soc - np.linspace(0, 360, n_soc + 1)
```

```python
# GREEN ZONE
thetas_ring = np.linspace(0, 360, 400)
theta_ring  = np.concatenate([thetas_ring, thetas_ring[::-1]])

r_outer_fill = np.full_like(thetas_ring, r_out_social_ring - 0.02)
r_inner_fill = np.full_like(thetas_ring, r_in_social_ring + 0.02)
r_vals_fill  = np.concatenate([r_outer_fill, r_inner_fill[::-1]])
fig.add_trace(go.Scatterpolar(
    r=r_vals_fill, theta=theta_ring, mode="lines",
    fill="toself",
    fillcolor="rgba(80,160,90,1)",
    line=dict(width=0),
    hoverinfo="skip"
))

r_outer = np.full_like(thetas_ring, r_out_social_ring)
r_inner = np.full_like(thetas_ring, r_out_social_ring - 0.18)
r_vals  = np.concatenate([r_outer, r_inner[::-1]])
fig.add_trace(go.Scatterpolar(
    r=r_vals, theta=theta_ring, mode="lines",
    fill="toself",
    fillcolor="rgba(12,90,40,1)",
    line=dict(width=0),
    hoverinfo="skip"
))

r_outer2 = np.full_like(thetas_ring, r_in_social_ring + 0.18)
r_inner2 = np.full_like(thetas_ring, r_in_social_ring)
r_vals2  = np.concatenate([r_outer2, r_inner2[::-1]])
fig.add_trace(go.Scatterpolar(
    r=r_vals2, theta=theta_ring, mode="lines",
    fill="toself",
    fillcolor="rgba(34,116,67,1)",
    line=dict(width=0),
    hoverinfo="skip"
))

# SOCIAL RING
for i in range(n_soc):
    theta1, theta2 = angles_soc[i], angles_soc[i + 1]
    thetas = np.linspace(theta1, theta2, 10)
    theta  = np.concatenate([thetas, thetas[::-1]])

    r_outer = np.full_like(thetas, r_out_arr_soc[i])
    r_inner = np.full_like(thetas, r_in_arr_soc[i])
    r_vals  = np.concatenate([r_outer, r_inner[::-1]])

    vn        = v_soc_norm[i]
    fillcolor = soc_to_color(vn)

    ind_name = soc_valid["indicator"].iloc[i]
    title    = social_titles_by_indicator.get(ind_name, ind_name)

    fig.add_trace(go.Scatterpolar(
        r=r_vals,
        theta=theta,
        mode="lines",
        fill="toself",
        fillcolor=fillcolor,
        line=dict(width=0),
        hoverinfo="text",
        text=(
            f"{title}: {soc_valid['last_value'].iloc[i]:.2f}<br>"
        )
    ))

# DIMENSIONS
dim_angles = {}
for i in range(n_soc):
    dim = soc_valid["dimension"].iloc[i]
    theta1, theta2 = angles_soc[i], angles_soc[i + 1]
    mid = (theta1 + theta2) / 2.0
    dim_angles.setdefault(dim, []).append(mid)

for dim, mids in dim_angles.items():
    theta_mid = np.mean(mids)

    words = dim.split()
    if len(words) > 2:
        words = [" ".join(words[:-1]), words[-1]]

    r_base = (r_in_social_ring + r_out_social_ring) / 2 + 0.05
    line_offsets = [0.0, -0.18]

    for j, w in enumerate(words):
        r_here = r_base + (line_offsets[j] if j < len(line_offsets) else 0)

        fig.add_trace(go.Scatterpolar(
            r=[r_here],
            theta=[theta_mid],
            mode="text",
            text=[w],
            textfont=dict(color="white", size=9, family="Times New Roman"),
            hoverinfo="skip"
        ))

# green rings
text_sf = "SOCIAL FOUNDATION"
n_sf = len(text_sf)
angles_sf = np.linspace(210, 330, n_sf)
r_text_sf = (r_out_social_ring + r_in_social_ring) / 2 - 0.3

fig.add_trace(go.Scatterpolar(
    r=[r_text_sf] * n_sf,
    theta=angles_sf,
    mode="text",
    text=list(text_sf),
    line=dict(width=0),
    textfont=dict(color="white", size=8, family="Times New Roman"),
    hoverinfo="skip"
))

text_ec = "ECOLOGICAL CEILING"
n_ec = len(text_ec)
angles_ec = np.linspace(150, 30, n_ec)
r_text_ec = r_in_eco + 0.20

fig.add_trace(go.Scatterpolar(
    r=[r_text_ec] * n_ec,
    theta=angles_ec,
    mode="text",
    text=list(text_ec),
    line=dict(width=0),
    textfont=dict(color="white", size=8, family="Times New Roman"),
    hoverinfo="skip"
))

# FIG LAYOUT
fig.update_layout(
    width=900,   # ή 1000
    height=900,
    hoverlabel=dict(
        bgcolor="white",
        font_color="black",
        bordercolor="black"
    ),
    showlegend=False,
    paper_bgcolor="white",
    plot_bgcolor="white",
    polar=dict(
        bgcolor="white",
        radialaxis=dict(visible=False, showline=False, showgrid=False,
                        range=[0, r_max_extra * 1.05]),
        angularaxis=dict(
            visible=False,
            showline=False,
            showgrid=False
        )
    ),
    margin=dict(l=0, r=0, t=0, b=0)
)

fig.show()
```
